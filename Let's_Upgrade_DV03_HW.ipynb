```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

=> loading the fmri dataset from seaborn.load_dataset()

```python
fmri = sns.load_dataset('fmri')
```

#Visualizing the dataset

```python
sns.relplot(data = fmri, x = 'timepoint', y = 'signal', hue = 'event', size = 'event', kind ='line')
plt.title('Relation b/w timepoint and signal')
plt.show()
```

```python
sns.relplot(data = fmri, x = 'timepoint', y = 'signal', hue= 'region')
plt.title('Relation b/w timepoint and signal')
plt.show()
```

From the above graph we see that,
1. most of the signal in both region lies b/w -0.2 to 0.2
2. In the above interval we have frontal region > parietal region

```python
sns.boxplot(data = fmri, x = 'timepoint', y = 'signal', hue = 'region')
plt.title('Relation b/w timepoint and signal')
plt.show()
```

The Above graph we can see that,
1. timepoint b/w 3 and 8 has more density than rest of the plot
2. There are some outliers present which can affect the model performance.
